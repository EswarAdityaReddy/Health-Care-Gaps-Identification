```python
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("final_dataset.csv")   # make sure file is in same folder
df.head()
```

```python
X = df['text'].astype(str)   # input text
y = df['label']              # target labels
```

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print(label_encoder.classes_)   # see label names
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
def tokenize(texts):
    return tokenizer(
        list(texts),
        padding=True,
        truncation=True,
        max_length=128,
        return_tensors="pt"
    )

train_encodings = tokenize(X_train)
test_encodings = tokenize(X_test)
```

```python
import torch

class HealthDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

train_dataset = HealthDataset(train_encodings, y_train)
test_dataset = HealthDataset(test_encodings, y_test)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=2,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="epoch",   # ✅ FIXED HERE
    logging_dir="./logs",
    save_strategy="no"
)
```

```python
from sklearn.metrics import accuracy_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = logits.argmax(axis=1)
    return {"accuracy": accuracy_score(labels, preds)}
```

```python
from transformers import BertForSequenceClassification

num_labels = len(set(y_encoded))

model = BertForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=num_labels
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)
```

```python
import torch
print("CUDA:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0))
```

```python
trainer.train()
```

```python
from sklearn.metrics import accuracy_score, classification_report

predictions = trainer.predict(test_dataset)
y_pred = predictions.predictions.argmax(axis=1)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def predict_gap(text):
    inputs = tokenizer(
        text,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=128
    )
    
    # 🔥 Move inputs to GPU
    inputs = {key: val.to(device) for key, val in inputs.items()}
    
    # Ensure model is also on GPU
    model.to(device)
    
    outputs = model(**inputs)
    logits = outputs.logits
    pred = torch.argmax(logits, dim=1).item()
    
    return label_encoder.inverse_transform([pred])[0]

print(predict_gap("There is shortage of nurses in rural hospitals"))
```
